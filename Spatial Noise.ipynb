```python
import numpy as np
from matplotlib import pyplot as plt

import timeit
```

```python
N = 2**(7)
print(N)
```

```python
def gaussian(x,y):
    return np.exp(-(x**2+ y**2)*1**2)
```

```python
x, y = np.meshgrid(np.linspace(-10,10,N), np.linspace(-10,10,N))
```

```python
gauss = gaussian(x,y)
```

```python
fgauss = np.abs(np.fft.fft2(gauss))  # use to mask
```

```python
plt.imshow(fgauss)
```

```python
fim = np.random.normal(size = (N,N))
im = np.abs(np.fft.ifft2(fim * fgauss))
```

```python
plt.imshow(np.abs(im)[0:int(N/4),0:int(N/4)])
plt.colorbar()
```

```python
plt.imshow(fim)
```

```python
plt.imshow(fim*fgauss)
```

```python
def f():
    fim = np.random.normal(size = (N,N))
    im = np.abs(np.fft.ifft2(fim * fgauss))
```

```python
timeit.timeit(f, number = 1000)
```
